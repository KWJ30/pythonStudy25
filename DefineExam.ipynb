```python
# 여러명의 데이터를 리스트로 입력받아 처리
# 성적에 대한 CRUD
# 학생에 대한 CRUD
# 통계에 대한 CRUD 등을 작성한다
# 함수를 활용해서 학생의 석차를 구하는 함수

# 성적을 등급으로 변환하는 함수
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# 성적을 등수로 변환하는 함수
def get_rank(score, scores):
    rank = 1
    for s in scores:
        if score < s:
            rank += 1
    return rank

# 동석차 학생 수 구하는 함수
def get_tied_rank_count(score, scores):
    tied_count = scores.count(score)
    return tied_count  # 동석차 학생 수를 반환

name, kor, eng, math, py = [],[],[],[],[]
subjects = [kor, eng, math, py]
subject_names = ["국어", "영어", "수학", "파이썬"]
total_scores = []
error_count = 0

# 초기값 설정
initial_data = [
    ("김철수", [85, 90, 80, 70]),
    ("이영희", [85, 92, 75, 85]),
    ("박민수", [78, 85, 85, 95]),
    ("정다은", [90, 85, 90, 92]),
    ("최준호", [75, 65, 90, 85])
]

# 초기값을 리스트에 추가
for input_name, scores in initial_data:
    if input_name not in name:
        name.append(input_name)
        for i, temp_score in enumerate(scores):
            subjects[i].append(temp_score)
        total_scores.append(sum(scores))


run = True
confirm = ""
prompt = """1. 이름/성적 입력
2. 성적 확인
3. 성적 수정
4. 성적 삭제(초기화)
5. 이름 수정/삭제
6. 프로그램 종료

Enter number(1~6): """

while run:
    print(prompt)
    select = input("번호 입력: ")

    if select == "1":
        print("이름 및 성적 입력\n")
        input_name = input("이름: ").strip()    # 좌우 공백 제거
        input_name = input_name.replace(" ", "")  # 모든 공백 제거
        if input_name in name:
            print("이미 등록된 이름입니다.")
            continue

        # 성적 입력을 위한 리스트 초기화
        temp_subjects = []
        valid_input = True  # 모든 입력이 유효한지 확인하는 플래그

        for i, subject in enumerate(subjects):
            try:
                score = int(input(f"{subject_names[i]}: "))
                if 0 <= score <= 100:  # 점수가 0~100 사이인지 확인
                    temp_subjects.append(score)
                else:
                    print("점수는 0에서 100 사이여야 합니다.")
                    valid_input = False
                    break  # 점수 범위 벗어나면 중단
            except ValueError:
                print("잘못된 값을 입력하셨습니다. 숫자를 입력해주세요.")
                valid_input = False
                break  # 숫자가 아니면 중단

        if valid_input:
            name.append(input_name)
            for i, temp_score in enumerate(temp_subjects):
                subjects[i].append(temp_score)  # 성적이 정상적이면 subjects에 추가
            total = sum(temp_subjects)
            total_scores.append(total)
            print("입력되었습니다.\n")
        else:
            print("성적 입력이 취소되었습니다. 다시 시도해주세요.\n")

    elif select == "2":
        print("성적 확인\n")
        search_name = input("이름 입력: ").strip()  # 좌우 공백 제거
        search_name = search_name.replace(" ", "")  # 모든 공백 제거
        if search_name in name:
            index = name.index(search_name)
            total = 0  # 총점 초기화
            for i, subject in enumerate(subjects):
                score = subject[index]
                total += score  # 총점 누적
                grade = get_grade(score)
                subject_rank = get_rank(score, subject)  # 각 과목의 등수 구하기
                tied_rank_count = get_tied_rank_count(score, subject)  # 동석차 학생 수 구하기

                 # 동석차가 1 이상일 때만 표시
                if tied_rank_count > 1:
                    print(f"{subject_names[i]}: {score} ({grade}) {subject_rank}({tied_rank_count})/{len(name)}", end="  ")
                else:
                    print(f"{subject_names[i]}: {score} ({grade}) {subject_rank}/{len(name)}", end="  ")

            avg = total / len(subjects)  # 평균 계산
            grade = get_grade(avg)
            rank = get_rank(total, total_scores)
            tied_rank_count = get_tied_rank_count(total, total_scores)

            # 동석차가 1 이상일 때만 표시
            if tied_rank_count > 1:
                print(f"\n총점: {total}  평균: {avg:.2f} ({grade}) 등수: {rank}({tied_rank_count})/{len(name)} \n")
            else:
                print(f"\n총점: {total}  평균: {avg:.2f} ({grade}) 등수: {rank}/{len(name)} \n")

        else:
            print("해당이름이 존재하지 않습니다.\n")

    elif select == "3":
        print("성적 수정\n")
        search_name = input("이름 입력: ")
        if search_name in name:
            index = name.index(search_name)
            for i, subject in enumerate(subjects):
                subject[index] = int(input(f"{subject_names[i]}: "))
            print("수정이 완료되었습니다.\n")
        else:
            print("해당이름이 존재하지 않습니다.\n")

    elif select == "4":
        print("성적 삭제(초기화)\n")
        while confirm not in ["y", "n"]:
            confirm = input("정말 삭제하시겠습니까? (y/n) ")
            if confirm == 'y':
                search_name = input("이름 입력: ")
                if search_name in name:
                    index = name.index(search_name)
                    for i, subject in enumerate(subjects):
                        subject[index]=0
                    print("삭제되었습니다.\n")
                else:
                    print("해당이름이 존재하지 않습니다.\n")

            elif confirm == 'n':
                print("삭제가 취소되었습니다.\n")
                confirm = ""
                break
            else:
                print("잘못된 입력입니다.\n")
                continue

    elif select == "5":
        print("이름 수정/삭제\n")
        name_select = int(input("1. 수정 2. 삭제: "))
        if name_select == 1:
            print("이름 수정\n")
            search_name = input("이름 입력: ")
            if search_name in name:
                index = name.index(search_name)
                name[index] = input("새로운 이름 입력: ")
                print("이름이 수정되었습니다.\n")
            else:
                print("해당이름이 존재하지 않습니다.\n")
        elif name_select == 2:
            print("이름 삭제\n")
            while confirm not in ["y", "n"]:
                confirm = input("정말 삭제하시겠습니까? (y/n) ")
                if confirm == 'y':
                    search_name = input("이름 입력: ")
                    if search_name in name:
                        index = name.index(search_name)
                        del name[index]
                        for subject in subjects:
                            del subject[index]
                        print("삭제되었습니다.\n")
                    else:
                        print("해당이름이 존재하지 않습니다.\n")

                elif confirm == 'n':
                    print("삭제가 취소되었습니다.\n")
                    confirm = ""
                    break

                else:
                    print("잘못된 입력입니다.\n")
                    continue

    elif select == "6":
        print("프로그램 종료")
        run = False

    else :
        error_count += 1
        print("올바른 값을 입력해주세요.")
        print("오류 3회시 프로그램 강제종료. 현재 %d회" %error_count)
        if error_count >= 3:
            print("프로그램을 종료합니다.")
            run = False
```
