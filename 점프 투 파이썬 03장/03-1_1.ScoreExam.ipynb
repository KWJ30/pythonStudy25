```python
# 성적처리용 프로그램을 만들어본다
# 1. 키보드로 개인정보를 입력 받는다.
# 2. 키보드로 성적을 입력 받는다. (국어, 영어, 수학, 파이썬)   0~100
# 3. 입력된 정보를 확인한다. -> 잘못된 정보를 수정한다.
# 4. 통계를 만들어 출력한다.
# 5. 등급을 출력하고 종료한다.  (90이상 A, 80이상 B, 70이상 C, 60이상 D, 그외 F)

def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

print("=" * 25)
print("개인정보를 입력해주세요")
print("=" * 25)

name = input("이름 : ")
age = input("나이 : ")
add = input("주소 : ")

print("=" * 25)
print("%s 님 환영합니다" %name)
print("성적을 입력해주세요")
print("=" * 25)

subjects = ["국어", "영어", "수학", "파이썬"]
scores = {}
modi = True
confirm = ""

while modi:
    for subject in subjects:
        while True:
            try:
                score = int(input(f"{subject}: "))
                if score < 0 or score > 100:
                    raise ValueError
                scores[subject] = score
                break
            except ValueError:
                print("0~100 사이 정수를 입력하세요.")

    print("=" * 45)
    print("성적을 확인합니다.")
    print(" ".join(f"{k}:{v} " for k, v in scores.items()))
    print("=" * 45)

    confirm = ""

    while confirm not in ["y", "n"]:
        confirm = input("입력하신 성적을 수정할까요? (y/n): ")
        if confirm == "y":
            print("성적수정을 진행합니다.")
        elif confirm == "n":
            print("성적수정을 취소합니다.")
            modi = False
        else :
            print("잘못된 입력입니다. 다시 입력해주세요.")

print("=" * 35)

confirm = ""
total= sum(scores.values())
avg = total / len(scores)

while confirm not in ["y", "n"]:
        confirm = input("총점과 평균을 출력할까요? (y/n): ")
        if confirm == "y":
            print("총점과 평균을 출력합니다.")
            print("총점 : %d \t 평균 : %.2f" %(total, avg))
        else :
            print("잘못된 입력입니다. 다시 입력해주세요.")

print("=" * 35)

confirm = ""

while confirm not in ["y", "n"]:
        confirm = input("각 과목의 등급을 출력할까요? (y/n): ")
        if confirm == "y":
            print("등급을 출력합니다.")
            for subject, score in scores.items():
                print(f"{subject}: {get_grade(score)}")
        else :
            print("잘못된 입력입니다. 다시 입력해주세요.")


print("=" * 25)
print("프로그램을 종료합니다.")
print("=" * 25)
```
