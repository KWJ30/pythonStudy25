```python
from genericpath import exists
# 1회용이 아닌 반복문을 활용해서 성적 처리 프로그램을 만들어 보자.
# 1. 성적의 CRUD를 구현한다.
# 2. 1번메뉴를 선택하면 성적을 입력한다.
# 3. 2번메뉴를 선택하면 성적을 확인한다.
# 4. 3번메뉴를 선택하면 성적을 수정한다.
# 5. 4번메뉴를 선택하면 성적을 삭제한다.
# 6. 5번메뉴를 선택하면 프로그램을 종료한다.

run = True
confirm = ""
prompt = """1. 성적입력
2. 성적확인
3. 성적수정
4. 성적삭제
5. 프로그램종료

Enter number(1~5): """

while run:
    print(prompt)
    select = int(input("번호 입력: "))

    if select == 1:
        print("성적입력\n")
        kor = int(input("국어: "))
        eng = int(input("영어: "))
        math = int(input("수학: "))
        py = int(input("파이썬: "))
        print("입력되었습니다.\n")

    elif select == 2:
        print("성적확인\n")
        try :
            print("국어 : %d \t 영어 : %d \t 수학 : %d \t 파이썬 : %d \n" %(kor, eng, math, py))
        except :
            print("성적이 존재하지 않습니다.")

    elif select == 3:
        print("성적수정\n")
        if 'kor' not in globals():
            print("성적이 존재하지 않습니다.\n")
            continue
        kor = int(input("국어: "))
        eng = int(input("영어: "))
        math = int(input("수학: "))
        py = int(input("파이썬: "))
        print("수정되었습니다.\n")

    elif select == 4:
        print("성적삭제\n")
        if 'kor' not in globals():
            print("성적이 존재하지 않습니다.\n")
            continue
        while confirm not in ["y", "n"]:
            confirm = input("정말 삭제하시겠습니까? (y/n) ")
            if confirm == 'y':
                del kor, eng, math, py
                print("삭제되었습니다.\n")
            elif confirm == 'n':
                print("취소되었습니다.\n")
                confirm = ""
                break
            else:
                print("잘못된 입력입니다.\n")
                continue

    elif select == 5:
        print("프로그램 종료")
        run = False
```
