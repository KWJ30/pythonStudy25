```python
# 여러명의 데이터를 리스트로 입력받아 처리
# 성적에 대한 CRUD
# 학생에 대한 CRUD
# 통계에 대한 CRUD 등을 작성한다

name, kor, eng, math, py = [],[],[],[],[]
subjects = [kor, eng, math, py]
subject_names = ["국어", "영어", "수학", "파이썬"]

run = True
confirm = ""
prompt = """1. 이름/성적 입력
2. 성적 확인
3. 성적 수정
4. 성적 삭제(초기화)
5. 이름 수정/삭제
6. 프로그램 종료

Enter number(1~6): """

while run:
    print(prompt)
    select = int(input("번호 입력: "))

    if select == 1:
        print("이름 및 성적 입력\n")
        name.append(input("이름: "))
        for i, subject in enumerate(subjects):
            score = int(input(f"{subject_names[i]}: "))
            subject.append(score)
        print("입력되었습니다.\n")

    elif select == 2:
        print("성적 확인\n")
        search_name = input("이름 입력: ")
        if search_name in name:
            index = name.index(search_name)
            total = 0  # 총점 초기화
            for i, subject in enumerate(subjects):
                score = subject[index]
                total += score  # 총점 누적
                print(f"{subject_names[i]}: {score}", end="  ")  # 가로로 출력되도록 end=" " 추가
            avg = total / len(subjects)  # 평균 계산
            print(f"\n총점: {total}  평균: {avg:.2f}\n")  # 총점과 평균 출력
        else:
            print("해당이름이 존재하지 않습니다.\n")

    elif select == 3:
        print("성적 수정\n")
        search_name = input("이름 입력: ")
        if search_name in name:
            index = name.index(search_name)
            for i, subject in enumerate(subjects):
                subject[index] = int(input(f"{subject_names[i]}: "))
            print("수정이 완료되었습니다.\n")
        else:
            print("해당이름이 존재하지 않습니다.\n")

    elif select == 4:
        print("성적 삭제(초기화)\n")
        while confirm not in ["y", "n"]:
            confirm = input("정말 삭제하시겠습니까? (y/n) ")
            if confirm == 'y':
                search_name = input("이름 입력: ")
                if search_name in name:
                    index = name.index(search_name)
                    for i, subject in enumerate(subjects):
                        subject[index]=0
                    print("삭제되었습니다.\n")
                else:
                    print("해당이름이 존재하지 않습니다.\n")

            elif confirm == 'n':
                print("삭제가 취소되었습니다.\n")
                confirm = ""
                break
            else:
                print("잘못된 입력입니다.\n")
                continue

    elif select == 5:
        print("이름 수정/삭제\n")
        name_select = int(input("1. 수정 2. 삭제: "))
        if name_select == 1:
            print("이름 수정\n")
            search_name = input("이름 입력: ")
            if search_name in name:
                index = name.index(search_name)
                name[index] = input("새로운 이름 입력: ")
                print("이름이 수정되었습니다.\n")
            else:
                print("해당이름이 존재하지 않습니다.\n")
        elif name_select == 2:
            print("이름 삭제\n")
            while confirm not in ["y", "n"]:
                confirm = input("정말 삭제하시겠습니까? (y/n) ")
                if confirm == 'y':
                    search_name = input("이름 입력: ")
                    if search_name in name:
                        index = name.index(search_name)
                        del name[index]
                        for subject in subjects:
                            del subject[index]
                        print("삭제되었습니다.\n")
                    else:
                        print("해당이름이 존재하지 않습니다.\n")

                elif confirm == 'n':
                    print("삭제가 취소되었습니다.\n")
                    confirm = ""
                    break

                else:
                    print("잘못된 입력입니다.\n")
                    continue

    elif select == 6:
        print("프로그램 종료")
        run = False
```
